```python
# Importing Libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

# ignoring warnings
import warnings
warnings.simplefilter("ignore")

# set max output to 10 rows
pd.set_option('max_rows', 10)
pd.set_option('max_columns', 30)

# reading in data
data = pd.read_csv('parkinsons.data')

# previewing data
data.head(10)
```

# 1.1 Creating Ensemble Classifiers

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import VotingClassifier
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC # can use linear or with kernel

# splitting data into X and y
y = data['status']
X = data.drop('status', axis = 1)

# 80/20 split of test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# defining the classifiers
clf1 = LogisticRegression()
clf2 = SVC()

# creating pipeline
pipeline = Pipeline([
    ('VC', VotingClassifier([
        ('LR', clf1),
        ('SVC', clf2)
    ]))
    
])

# defining parameters
parameters = {
    'VC__LR__penalty':('l1', 'l2'), # l2
    'VC__LR__C': [0.5, 1, 1.5], # 0.5
    'VC__LR__max_iter': [50,100,150], # 50
    'VC__SVC__kernel':('rbf', 'linear'), # rbf
    'VC__SVC__max_iter':[50,100], # 100
    'VC__SVC__C': [0.5, 1, 1.5] # 1.5
}

# setting up gridsearch
clf3 = GridSearchCV(pipeline, parameters, verbose=1, scoring='f1')
clf3.fit(X_train,y_train)

# printing out results
print('Best score and parameter combination = ')
print(clf3.best_score_)    
print(clf3.best_params_)  
```

### Comments (TALK ABOUT DIFFERENT PARAMETERS)

# 1.2 Comparing F1 Scores

```python
from sklearn.metrics import confusion_matrix, f1_score

# creating classifiers
clf4 = LogisticRegression(penalty= 'l2', C = 0.5, max_iter= 50)
clf5 = SVC(kernel= 'rbf', max_iter= 100, C = 1.5)

# fitting classifiers
clf4.fit(X_train,y_train)
clf5.fit(X_train,y_train)

# creating confusion matricies
CF1 = confusion_matrix(y_test, clf4.predict(X_test))
CF2 = confusion_matrix(y_test, clf5.predict(X_test))
CF3 = confusion_matrix(y_test, clf3.predict(X_test))

# plotting the confusion matrices
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize = (10,20))
sns.heatmap(CF1, annot=True, fmt = '.4g', cmap = 'plasma',ax = ax1);
ax1.set_title('Logistic Regression', fontsize = 20)
ax1.set_xlabel('Class predicted by model')
ax1.set_ylabel('Actual class')
sns.heatmap(CF2, annot=True, fmt = '.4g', cmap = 'plasma',ax = ax2);
ax2.set_title('Support Vector Machine', fontsize = 20)
ax2.set_xlabel('Class predicted by model')
ax2.set_ylabel('Actual class')
sns.heatmap(CF3, annot=True, fmt = '.4g', cmap = 'plasma',ax = ax3);
ax3.set_title('Voting Classifier', fontsize = 20)
ax3.set_xlabel('Class predicted by model')
ax3.set_ylabel('Actual class')
```

```python
# printing out f1 scores
print('Logistic Regression f1 score: '+ str(round(100*f1_score(y_test,clf4.predict(X_test) ),2)) + '%')
print('Support Vecotor Machine f1 score: '+ str(round(100*f1_score(y_test,clf5.predict(X_test) ),2)) + '%')
print('Voting Classifier f1 score: '+ str(round(100*f1_score(y_test,clf3.predict(X_test) ),2)) + '%')
```
